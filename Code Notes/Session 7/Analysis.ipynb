# Step 04: Data Cleaning

- Date ko datetime format me convert karna zaroori hai taake trend analysis ho sake.

```python
df["Date"] = pd.to_datetime(df["Date"])
```

# Step 05: Data Visualization

# 1. Sales Distribution
- Histogram batata hai data ka spread kaisa hai.

Insight:

- Most transactions medium range me hain

-  high value sales bhi hain

```python
plt.figure(figsize = (8,5))
sns.histplot(df["Sales"], bins=30,kde=True)
plt.title("Sales Distribution")
plt.show()
```

# 2.Branch-wise Sales

Purpose:

- Kaunsi branch zyada revenue generate karti hai?

Insight:
- Branch comparison clear ho jata hai.

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Branch", y="Sales", data=df)
plt.title("Branch Wise Sale")
plt.show()
```

# 03. Product Line Performance

Purpose:

- Kaunsa product sabse zyada bikta hai?

```python
plt.Figure(figsize=(10,5))
sns.barplot(x="Product line", y="Sales", data=df)
plt.xticks(rotation=45)
plt.title("Sales by Product Line")
plt.show()
```

# 4. Gender Distribution
Purpose:

- Male ya Female customers zyada?

```python
sns.countplot(x= "Gender", data=df)
plt.title("Customer Gender Distribution ")
plt.show()
```

# 5. Payment Method Analysis
Purpose:

- ustomers ka preferred payment method kya hai?

```python
sns.countplot(x="Payment", data=df)
plt.title("Payment Method Usage")
plt.show()
```

# 6. Daily Sales Trend
Purpose:

- Time ke sath sales ka trend.
- Groupby Date kar ke har din ki total sales calculate ki.

```python
daily_sales = df.groupby("Date")["Sales"].sum()

plt.figure(figsize=(10,5))
daily_sales.plot()
plt.title("Daily Sales Trend")
plt.show()
```

# 7. Sales vs Gross Income Relationship

Purpose:

- Zyada sales = zyada profit?

```python
sns.scatterplot(x="Sales", y="gross income", data=df)
plt.title("Sales vs Gross Income")
plt.show()
```

# 8. Correlation Heatmap

Purpose:
- Variables ke beech relationship check karna.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
