# STEP 8: Single Column Analysis

- Is se hum sirf house prices ka analysis karte hain:

```python
df['medv'].describe()
```

# STEP 9: Data Filtering (Conditions)

## Example: Mehngi Houses

Explanation

- Ye sirf wo houses dikhata hai jinki price 30 se zyada hai

- Condition based analysis ke liye use hota hai

```python
df[df['medv'] > 30]
```

# Multiple Conditions
## Explanation:

- Zyada rooms

- Zyada price

Real estate analysis me bohat useful.

```python
df[(df["medv"] < 30) & (df["rm"] > 7)]
```

# STEP 10: Sorting Data

Houses by Price (High to Low)

## Explanation

Highest price wale houses upar aa jate hain

Ranking aur comparison ke liye use hota hai

```python
df.sort_values("medv", ascending=False)
```

# STEP 11: Correlation Analysis (IMPORTANT)

## Explanation

Correlation batata hai:

- Kaun se factors price ko increase ya decrease karte hain

Key findings:

- rm (rooms) → strong positive relation

- lstat → strong negative relation

Matlab:

- Zyada rooms = zyada price
- Lower status zyada = kam price

```python
df.corr()["medv"].sort_values(ascending= False)
```

# STEP 12: Groupby Analysis
- River ke paas houses ka price

## Explanation

- chas = 1 → river ke paas

- chas = 0 → river se door

River ke paas houses ki average price zyada hoti hai.

```python
df.groupby("chas")["medv"].mean()
```

# STEP 13: New Feature Banana

## Explanation

- New categorical column banaya

- ML aur reporting me bohat use hota hai


```python
df["Price_Level"] = df["medv"].apply(
    lambda x: "High" if x > 30 else "Normal"
)
```

# STEP 14: Clean Dataset Save Karna
## Explanation

- Cleaned / analyzed data ko save kiya

- Future use ke liye ready

```python
df.to_csv("BostonHousing_Clean.csv", index=False)
```
